```
def encode(val):
  count = val.count('1')
  x = val.split("1")
  #print("count of 1s",count)
  for i in range(count):
    if x[i] == '':
      #print("true")
      print("00",end=" ")
    else:
      #print("No of Zeros: "+ str(len(x[i])))
      bits = math.ceil(math.log(len(x[i]), 2))
      #print("bits Required to represent, ", bits)
      j = [1 for m in range(bits -1)]
      for n in j:
        print (n,end="")
      print(end = ' ')
      #print(j)
      print('0',end = " ")
      print(decimalToBinary(len(x[i])), end = " ")

```
